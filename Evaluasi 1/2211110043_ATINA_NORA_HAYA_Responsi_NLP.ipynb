##### Soal 2
Lakukan pembersihan meliputi penghapusan punctuation, angka dan karakter yang
tidak penting menggunakan menggunakan regex.
Simpan hasilnya menjadi file csv/excel.

```python
data = data['content'].to_list()
data
```

```python
import re
import string

def menghilangkan_tanda_baca_dan_angka(data):
  for char in data:
    if char in string.punctuation or char.isdigit():
      data = data.replace(char, '')
  return data
data_clean = [menghilangkan_tanda_baca_dan_angka(data) for data in data]
data = pd.DataFrame(data_clean)
data
```

```python
data.to_csv('soal2.csv')
```

##### Soal 3D
Lakukan stopword removing berdasarkan hasil 3c, simpan hasilnya dalam
bentuk csv/excel.

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words('indonesian'))
stopwords
```

```python
data_clean = [word for word in tokens if word.lower() not in stopwords]
data_clean
```

```python
df = pd.DataFrame(data_clean)
df.to_csv('soal3d.csv')
```
